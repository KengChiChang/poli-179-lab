```python
import pandas as pd

news_df = pd.read_json("../data/News_Category_Dataset_v3.json", lines=True)
news_df
```

```python
category_select = ["POLITICS", "ENTERTAINMENT", "BUSINESS", "SPORTS", "TRAVEL"]
news_df = news_df.query('category == @category_select & short_description != "" & headline != "" & date >= "2016-01-01"')
news_df
```

```python
news_df['category'].value_counts()
```

```python
news_df['text'] = news_df['headline'] + '  |||  ' + news_df['short_description']
```

```python
news_df = news_df[['category', 'text']]
```

```python
news_df_sample = pd.concat([
    news_df.query('category == "POLITICS"').sample(frac=500/25500, random_state=123),
    news_df.query('category == "ENTERTAINMENT"').sample(frac=500/9600, random_state=123),
    news_df.query('category == "TRAVEL"').sample(frac=500/700, random_state=123),
    news_df.query('category == "BUSINESS"').sample(frac=500/1350, random_state=123),
    news_df.query('category == "SPORTS"').sample(frac=500/2250, random_state=123)
])
```

```python
news_df_sample['category'].value_counts()
```

```python
news_df_sample['text'] = news_df_sample['text'].str.replace("â€™", "'")
```

```python
news_df_sample.to_csv("../data/news_sample.csv", index=False, quoting=1)
```
